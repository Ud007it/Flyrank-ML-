## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

I'll choose scoring/ranking
My lane produces a ranked list of pages that need a content refresh. This isn't classification with many categories, and it isn't clustering (I'm not grouping similar pages together). It's a scoring task — the model gives each page a priority score, and pages get sorted (ranked) by that score so the content team knows what to work on first.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

Target: is_declining_label

## 3. Success metric

*One metric you can defend. What number means 'good'?*

Success metric: Precision@50
A content team can only realistically review about 50 pages a week. So the metric that matters isn't overall accuracy across all 30,000 pages — it's: of the top 50 pages my ranking flags, how many are actually declining? That's Precision@50.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

Why ML beats a fixed rule here
My hand-written rule assumed the most important signal was staleness combined with visibility (days_since_last_update >= 180 AND impressions_90d >= 500). But when I trained a decision tree, it split first on impressions_90d, then on content_age_days — not on days_since_last_update at all. This shows the real pattern in the data isn't the one I intuitively guessed. A fixed if/else rule only works if a human happens to pick the right combination of conditions upfront; a model can search across many features and find the actual signal, even when it's different from what a person would assume.
