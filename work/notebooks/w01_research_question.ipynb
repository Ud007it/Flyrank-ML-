```python
import pandas as pd


df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

print("Dataset Shape:", df.shape)
print("-" * 30)
print("Average Impressions (90 Days):", round(df["impressions_90d"].mean(), 2))
print("Average Content Age (Days):", round(df["content_age_days"].mean(), 2))
print("Average Sessions (90 Days):", round(df["sessions_90d"].mean(), 2))
print("-" * 30)
print("Trend Direction Distribution:")
print(df["trend_direction"].value_counts())
```
